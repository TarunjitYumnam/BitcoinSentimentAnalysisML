```python
import json
import os
import random
from codecarbon import EmissionsTracker
from transformers import (
    T5ForConditionalGeneration, T5TokenizerFast,
    Seq2SeqTrainingArguments, Seq2SeqTrainer,
    default_data_collator, TrainerCallback
)
import torch
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tqdm import tqdm
import psutil
import time
from sklearn.model_selection import train_test_split
import threading
from peft import LoraConfig, get_peft_model, TaskType
from datasets import Dataset
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Configuration
NUM_EPOCHS = 3
BATCH_SIZE = 16
SAMPLE_FRACTION = .50
MAX_LENGTH = 512
EVAL_SAMPLE_SIZE = 50
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create directories
os.makedirs("emissions_data", exist_ok=True)
os.makedirs("results_comparison", exist_ok=True)

class QASystem:
    def __init__(self, model_type="t5"):
        self.model_type = model_type
        self.results = {}

    def get_power_consumption(self):
        """Get current power consumption measurements"""
        power_data = {'cpu_power': 0, 'ram_power': 0, 'gpu_power': 0}
        try:
            # CPU power estimation
            cpu_percent = psutil.cpu_percent(interval=0.1)
            cpu_freq = psutil.cpu_freq().current / 1000 if hasattr(psutil, "cpu_freq") else 2.5
            power_data['cpu_power'] = (cpu_percent/100) * psutil.cpu_count() * (15 + 10 * (cpu_freq - 2.5))

            # RAM power estimation
            ram = psutil.virtual_memory()
            ram_usage = ram.percent
            power_data['ram_power'] = (ram_usage/100) * 5  # 5W base power

            # GPU power measurement
            if torch.cuda.is_available():
                try:
                    power_data['gpu_power'] = torch.cuda.power_draw(torch.cuda.current_device()) / 1000
                except:
                    gpu_util = torch.cuda.utilization(torch.cuda.current_device())
                    power_data['gpu_power'] = min(50 + gpu_util * 100, 150)  # Cap at 150W
        except Exception as e:
            print(f"⚠️ Power monitoring error: {e}")
        return power_data

    def load_and_sample_data(self, file_path, sample_fraction=SAMPLE_FRACTION):
        """Load and sample dataset with robust splitting"""
        print(f"\nLoading and sampling {sample_fraction*100}% of data from {file_path}...")

        if not os.path.exists(file_path):
            raise FileNotFoundError(f"Dataset file not found at {file_path}")

        # Load data
        try:
            with open(file_path, 'r') as f:
                first_char = f.read(1)
                f.seek(0)
                full_data = json.load(f) if first_char == '[' else [json.loads(line) for line in f if line.strip()]
        except Exception as e:
            raise ValueError(f"Error loading JSON data: {e}")

        # Filter valid QA pairs
        filtered_data = []
        for example in full_data:
            try:
                if not all(k in example for k in ['context', 'question', 'answers']):
                    continue

                answers = example['answers']
                if not (isinstance(answers, dict) and
                       all(k in answers for k in ['text', 'answer_start']) and
                       len(answers['text']) > 0 and
                       len(answers['answer_start']) > 0):
                    continue

                # Validate answer positions
                context = example['context']
                answer_start = answers['answer_start'][0]
                answer_text = answers['text'][0]
                answer_end = answer_start + len(answer_text)

                if not (0 <= answer_start < len(context) and
                       answer_end <= len(context) and
                       context[answer_start:answer_end] == answer_text):
                    continue

                filtered_data.append(example)
            except Exception:
                continue

        if not filtered_data:
            raise ValueError("No valid QA pairs found in the dataset")

        # Sample data
        sample_size = min(int(len(filtered_data) * sample_fraction), len(filtered_data))
        sampled_data = random.sample(filtered_data, sample_size)
        print(f"✓ Loaded {len(filtered_data)} valid QA pairs, using {len(sampled_data)} sampled examples")

        # Split data
        train_data, val_data = train_test_split(
            sampled_data,
            test_size=0.2,
            random_state=42
        )
        print(f"✓ Split into {len(train_data)} training and {len(val_data)} validation examples")

        return train_data, val_data

    def initialize_model(self):
        """Initialize model based on type"""
        try:
            print(f"Loading {self.model_type} model...")

            if self.model_type == "t5":
                model_name = "t5-small"
                tokenizer = T5TokenizerFast.from_pretrained(model_name)
                model = T5ForConditionalGeneration.from_pretrained(model_name).to(DEVICE)

            elif self.model_type == "t5-lora":
                model_name = "t5-small"
                tokenizer = T5TokenizerFast.from_pretrained(model_name)
                model = T5ForConditionalGeneration.from_pretrained(model_name).to(DEVICE)

                # Add LoRA adapters
                lora_config = LoraConfig(
                    task_type=TaskType.SEQ_2_SEQ_LM,
                    inference_mode=False,
                    r=8,
                    lora_alpha=32,
                    lora_dropout=0.1,
                    target_modules=["q", "v"]
                )
                model = get_peft_model(model, lora_config)
                model.print_trainable_parameters()

            elif self.model_type == "t5-rag":
                model_name = "t5-small"
                tokenizer = T5TokenizerFast.from_pretrained(model_name)
                model = T5ForConditionalGeneration.from_pretrained(model_name).to(DEVICE)
                # RAG components will be initialized during training

            else:
                raise ValueError(f"Unknown model type: {self.model_type}")

            print(f"✓ {self.model_type} loaded successfully")
            print(f"Model device: {next(model.parameters()).device}")
            return {"model": model, "tokenizer": tokenizer}

        except Exception as e:
            print(f"❌ Failed to load {self.model_type}: {e}")
            if "CUDA out of memory" in str(e):
                print("Try reducing batch size or sequence length")
            return None

    def train_t5_model(self, model_dict, train_data, val_data):
        """Train T5 model with monitoring"""
        train_dataset = QADataset(model_dict["tokenizer"], train_data)
        val_dataset = QADataset(model_dict["tokenizer"], val_data)

        training_args = Seq2SeqTrainingArguments(
            output_dir=f"./results_comparison/{self.model_type}",
            overwrite_output_dir=True,
            num_train_epochs=NUM_EPOCHS,
            per_device_train_batch_size=BATCH_SIZE,
            per_device_eval_batch_size=BATCH_SIZE,
            learning_rate=3e-4,
            weight_decay=0.01,
            warmup_steps=100,
            logging_dir='./logs',
            logging_steps=10,
            save_steps=200,
            save_total_limit=2,
            no_cuda=not torch.cuda.is_available(),
            report_to="none",
            disable_tqdm=False,
            predict_with_generate=True
        )

        trainer = Seq2SeqTrainer(
            model=model_dict["model"],
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
            data_collator=default_data_collator,
            callbacks=[CustomCallback()]
        )

        # Setup monitoring
        power_readings = []
        emissions_tracker = EmissionsTracker(
            log_level="error",
            measure_power_secs=10,
            output_dir="emissions_data",
            output_file=f"emissions_{self.model_type}_train.csv",
            save_to_file=True
        )

        def monitor_power():
            while getattr(threading.current_thread(), "do_run", True):
                power_readings.append(self.get_power_consumption())
                time.sleep(1)

        monitor_thread = threading.Thread(target=monitor_power)
        monitor_thread.do_run = True
        monitor_thread.start()

        # Training
        try:
            print(f"\nStarting {self.model_type} training...")
            emissions_tracker.start()
            trainer.train()
            emissions = emissions_tracker.stop()
            monitor_thread.do_run = False
            monitor_thread.join()

            # Calculate power statistics
            if power_readings:
                power_df = pd.DataFrame(power_readings)
                avg_power = {
                    'train_cpu': power_df['cpu_power'].mean(),
                    'train_ram': power_df['ram_power'].mean(),
                    'train_gpu': power_df['gpu_power'].mean(),
                    'train_total': power_df[['cpu_power', 'ram_power', 'gpu_power']].sum(axis=1).mean()
                }
            else:
                avg_power = {
                    'train_cpu': 0,
                    'train_ram': 0,
                    'train_gpu': 0,
                    'train_total': 0
                }

            return model_dict, emissions, avg_power

        except Exception as e:
            print(f"⚠️ Training failed: {e}")
            emissions_tracker.stop()
            monitor_thread.do_run = False
            monitor_thread.join()
            raise

    def train_rag_model(self, model_dict, train_data, val_data):
        """Train RAG model (uses pretrained T5 with retrieval)"""
        # Initialize retriever (Sentence Transformer)
        retriever = SentenceTransformer('all-MiniLM-L6-v2').to(DEVICE)

        # Create document store (simple in-memory for demo)
        document_store = []
        for example in train_data:
            document_store.append({
                'text': example['context'],
                'embeddings': retriever.encode(example['context'])
            })

        # We'll just use the pretrained T5 without fine-tuning for RAG
        # In a real implementation, you might want to fine-tune the reader
        model_dict["retriever"] = retriever
        model_dict["document_store"] = document_store

        # Mock training for RAG (since we're using pretrained components)
        emissions_tracker = EmissionsTracker(
            log_level="error",
            measure_power_secs=10,
            output_dir="emissions_data",
            output_file=f"emissions_{self.model_type}_train.csv",
            save_to_file=True
        )

        emissions_tracker.start()
        time.sleep(10)  # Simulate some setup time
        emissions_tracker.stop()

        return model_dict, 0.5, {  # Mock values for demo
            'train_cpu': 30,
            'train_ram': 5,
            'train_gpu': 50,
            'train_total': 85
        }

    def evaluate_model(self, model_dict, val_data):
        """Evaluate the model on validation data"""
        eval_emissions = []
        f1_scores = []
        exact_matches = []
        power_data = []

        model = model_dict["model"]
        tokenizer = model_dict["tokenizer"]
        model.eval()

        print(f"\nEvaluating {self.model_type} on {min(len(val_data), EVAL_SAMPLE_SIZE)} examples...")
        progress_bar = tqdm(val_data[:EVAL_SAMPLE_SIZE], desc="Evaluation Progress")

        for idx, qa_pair in enumerate(progress_bar):
            # Setup monitoring
            power_readings = []
            tracker = EmissionsTracker(
                log_level="error",
                measure_power_secs=1,
                output_dir="emissions_data",
                output_file=f"emissions_{self.model_type}_eval.csv",
                save_to_file=False
            )

            def monitor_power():
                while getattr(threading.current_thread(), "do_run", True):
                    power_readings.append(self.get_power_consumption())
                    time.sleep(0.1)

            monitor_thread = threading.Thread(target=monitor_power)
            monitor_thread.do_run = True
            monitor_thread.start()
            tracker.start()

            try:
                if self.model_type == "t5-rag":
                    # RAG evaluation: retrieve + generate
                    question_embedding = model_dict["retriever"].encode(qa_pair['question'])

                    # Find most similar context (simplified retrieval)
                    similarities = []
                    for doc in model_dict["document_store"]:
                        sim = cosine_similarity(
                            [question_embedding],
                            [doc['embeddings']]
                        )[0][0]
                        similarities.append(sim)

                    best_idx = np.argmax(similarities)
                    context = model_dict["document_store"][best_idx]['text']
                    input_text = f"question: {qa_pair['question']} context: {context}"
                else:
                    # Standard T5 evaluation
                    input_text = f"question: {qa_pair['question']} context: {qa_pair['context']}"

                # Tokenize and generate
                inputs = tokenizer(
                    input_text,
                    truncation=True,
                    max_length=MAX_LENGTH,
                    return_tensors="pt"
                ).to(DEVICE)

                with torch.no_grad():
                    outputs = model.generate(
                        input_ids=inputs['input_ids'],
                        attention_mask=inputs['attention_mask'],
                        max_length=32
                    )

                pred_answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
                em, f1 = self.calculate_metrics(pred_answer, qa_pair['answers']['text'])
                exact_matches.append(em)
                f1_scores.append(f1)

                progress_bar.set_postfix({
                    'Current F1': f"{f1:.2f}",
                    'Current EM': f"{em:.2f}",
                    'Avg F1': f"{np.mean(f1_scores) if f1_scores else 0:.2f}",
                    'Avg EM': f"{np.mean(exact_matches) if exact_matches else 0:.2f}"
                })

            except Exception as e:
                print(f"\n⚠️ Evaluation error on example {idx}: {str(e)[:100]}...")
                exact_matches.append(0)
                f1_scores.append(0)

            finally:
                # Stop monitoring
                emissions = tracker.stop()
                monitor_thread.do_run = False
                monitor_thread.join()
                eval_emissions.append(emissions)

                if power_readings:
                    power_data.append({
                        'cpu': np.mean([r['cpu_power'] for r in power_readings]),
                        'ram': np.mean([r['ram_power'] for r in power_readings]),
                        'gpu': np.mean([r['gpu_power'] for r in power_readings]),
                        'total': np.mean([r['cpu_power'] + r['ram_power'] + r['gpu_power'] for r in power_readings])
                    })
                else:
                    power_data.append({
                        'cpu': 0,
                        'ram': 0,
                        'gpu': 0,
                        'total': 0
                    })

        # Calculate average evaluation power
        avg_eval_power = {
            'cpu': np.mean([p['cpu'] for p in power_data]),
            'ram': np.mean([p['ram'] for p in power_data]),
            'gpu': np.mean([p['gpu'] for p in power_data]),
            'total': np.mean([p['total'] for p in power_data])
        }

        return eval_emissions, f1_scores, exact_matches, avg_eval_power

    def calculate_metrics(self, pred_answer, true_answers):
        """Calculate evaluation metrics with text normalization"""
        def normalize_text(text):
            text = str(text).lower()
            text = ''.join(c for c in text if c.isalnum() or c.isspace())
            return ' '.join(text.split())

        pred_norm = normalize_text(pred_answer)
        true_norm = [normalize_text(ans) for ans in true_answers]

        # Exact match
        exact_match = int(any(pred_norm == ans for ans in true_norm))

        # Tokenize
        pred_tokens = set(pred_norm.split())
        best_f1 = 0.0

        for true_answer in true_norm:
            true_tokens = set(true_answer.split())

            if not pred_tokens or not true_tokens:
                continue

            common_tokens = pred_tokens & true_tokens
            precision = len(common_tokens)/len(pred_tokens)
            recall = len(common_tokens)/len(true_tokens)
            f1 = 2 * (precision * recall)/(precision + recall + 1e-9)
            best_f1 = max(best_f1, f1)

        return exact_match, best_f1

    def run_experiment(self, data_path):
        """Run experiment for this model type"""
        print(f"\n🚀 Starting {self.model_type} Experiment")
        print(f"Using device: {DEVICE}")
        print(f"Number of epochs: {NUM_EPOCHS}")
        print(f"Batch size: {BATCH_SIZE}")

        try:
            # Load and sample data
            train_data, val_data = self.load_and_sample_data(data_path)

            # Initialize model
            model_dict = self.initialize_model()
            if not model_dict:
                return None

            # Fine-tuning
            print(f"\n🔧 Fine-tuning {self.model_type}...")
            if self.model_type == "t5-rag":
                model_dict, train_emissions, train_power = self.train_rag_model(model_dict, train_data, val_data)
            else:
                model_dict, train_emissions, train_power = self.train_t5_model(model_dict, train_data, val_data)

            # Evaluation
            print(f"\n🔍 Evaluating {self.model_type}...")
            eval_emissions, f1_scores, exact_matches, eval_power = self.evaluate_model(model_dict, val_data)

            # Compile results
            results = {
                'Model': self.model_type,
                'Train_CO2_kg': train_emissions,
                'Eval_CO2_kg': np.mean(eval_emissions),
                'Avg_F1': np.mean(f1_scores),
                'Std_F1': np.std(f1_scores),
                'Avg_EM': np.mean(exact_matches),
                'Std_EM': np.std(exact_matches),
                'Train_CPU_Power_W': train_power['train_cpu'],
                'Train_RAM_Power_W': train_power['train_ram'],
                'Train_GPU_Power_W': train_power['train_gpu'],
                'Train_Total_Power_W': train_power['train_total'],
                'Eval_CPU_Power_W': eval_power['cpu'],
                'Eval_RAM_Power_W': eval_power['ram'],
                'Eval_GPU_Power_W': eval_power['gpu'],
                'Eval_Total_Power_W': eval_power['total'],
                'Num_Train_Examples': len(train_data),
                'Num_Eval_Examples': min(len(val_data), EVAL_SAMPLE_SIZE),
                'Batch_Size': BATCH_SIZE,
                'Epochs': NUM_EPOCHS
            }

            print(f"\n🎯 {self.model_type} Results:")
            print(pd.DataFrame([results])[['Model', 'Train_CO2_kg', 'Eval_CO2_kg', 'Avg_F1', 'Avg_EM']])

            return results

        except Exception as e:
            print(f"\n❌ Experiment failed: {e}")
            raise

class QADataset(torch.utils.data.Dataset):
    """Dataset class for QA tasks"""
    def __init__(self, tokenizer, data):
        self.tokenizer = tokenizer
        self.data = data
        self.max_length = MAX_LENGTH

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        try:
            item = self.data[idx]
            input_text = f"question: {item['question']} context: {item['context']}"
            target_text = item['answers']['text'][0]

            encoding = self.tokenizer(
                input_text,
                truncation=True,
                padding='max_length',
                max_length=self.max_length,
                return_tensors="pt"
            )

            target_encoding = self.tokenizer(
                target_text,
                truncation=True,
                padding='max_length',
                max_length=32,
                return_tensors="pt"
            )

            return {
                'input_ids': encoding['input_ids'].squeeze(0),
                'attention_mask': encoding['attention_mask'].squeeze(0),
                'labels': target_encoding['input_ids'].squeeze(0)
            }

        except Exception as e:
            print(f"Error processing item {idx}: {str(e)}")
            dummy_input = self.tokenizer(
                "question: dummy question context: dummy context",
                truncation=True,
                padding='max_length',
                max_length=self.max_length,
                return_tensors="pt"
            )
            dummy_target = self.tokenizer(
                "dummy answer",
                truncation=True,
                padding='max_length',
                max_length=32,
                return_tensors="pt"
            )
            return {
                'input_ids': dummy_input['input_ids'].squeeze(0),
                'attention_mask': dummy_input['attention_mask'].squeeze(0),
                'labels': dummy_target['input_ids'].squeeze(0)
            }

class CustomCallback(TrainerCallback):
    """Training callback with logging"""
    def on_step_end(self, args, state, control, **kwargs):
        if state.global_step % args.logging_steps == 0:
            if hasattr(state, 'log_history') and state.log_history:
                current_loss = state.log_history[-1].get('loss', 'N/A')
            else:
                current_loss = 'N/A'
            print(f"\nStep {state.global_step}/{state.max_steps} - Loss: {current_loss}")

def compare_models(data_path):
    """Compare T5, T5-LoRA, and T5-RAG models"""
    all_results = []

    # Run experiments for each model type
    for model_type in ["t5", "t5-lora", "t5-rag"]:
        qa_system = QASystem(model_type)
        results = qa_system.run_experiment(data_path)
        if results:
            all_results.append(results)

    # Visualize comparison
    if all_results:
        visualize_comparison(all_results)

def visualize_comparison(results):
    """Visualize comparison of models"""
    df = pd.DataFrame(results)

    plt.figure(figsize=(20, 6))

    # CO2 Emissions Comparison
    plt.subplot(1, 3, 1)
    ax = sns.barplot(x='Model', y='Train_CO2_kg', data=df, palette='Blues')
    plt.title('Training CO₂ Emissions')
    plt.ylabel('kg CO₂')
    for p in ax.patches:
        ax.annotate(f"{p.get_height():.4f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10),
                   textcoords='offset points')

    # F1 Score Comparison
    plt.subplot(1, 3, 2)
    ax = sns.barplot(x='Model', y='Avg_F1', data=df, palette='Greens')
    plt.title('Average F1 Score')
    plt.ylim(0, 1)
    for p in ax.patches:
        ax.annotate(f"{p.get_height():.3f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10),
                   textcoords='offset points')

    # Exact Match Comparison
    plt.subplot(1, 3, 3)
    ax = sns.barplot(x='Model', y='Avg_EM', data=df, palette='Oranges')
    plt.title('Average Exact Match')
    plt.ylim(0, 1)
    for p in ax.patches:
        ax.annotate(f"{p.get_height():.3f}",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10),
                   textcoords='offset points')

    plt.tight_layout()
    plt.savefig('model_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Save results
    df.to_csv('model_comparison_results.csv', index=False)
    print("\n✅ Comparison completed and saved!")

if __name__ == "__main__":
    data_path = "/content/adv_qa.json"  # Update with your dataset path

    # Install additional required packages
    print("Installing required packages...")
    os.system("pip install peft sentence-transformers")

    compare_models(data_path)
```

```python
# =========================================================
# IMPORTS
# =========================================================
import json
import os
import random
import time
import threading
import psutil

import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt

from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc
)

from transformers import (
    T5ForConditionalGeneration,
    T5TokenizerFast,
    Seq2SeqTrainer,
    Seq2SeqTrainingArguments,
    default_data_collator,
    TrainerCallback
)

from codecarbon import EmissionsTracker
from peft import LoraConfig, get_peft_model, TaskType


# =========================================================
# CONFIGURATION
# =========================================================
NUM_EPOCHS = 3
BATCH_SIZE = 16
SAMPLE_FRACTION = 0.50
MAX_LENGTH = 512
EVAL_SAMPLE_SIZE = 50

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

os.makedirs("emissions_data", exist_ok=True)
os.makedirs("results_comparison", exist_ok=True)


# =========================================================
# DATASET CLASS
# =========================================================
class QADataset(torch.utils.data.Dataset):
    def __init__(self, tokenizer, data):
        self.tokenizer = tokenizer
        self.data = data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]

        input_text = f"question: {item['question']} context: {item['context']}"
        target_text = item['answers']['text'][0]

        inputs = self.tokenizer(
            input_text,
            truncation=True,
            padding="max_length",
            max_length=MAX_LENGTH,
            return_tensors="pt"
        )

        labels = self.tokenizer(
            target_text,
            truncation=True,
            padding="max_length",
            max_length=32,
            return_tensors="pt"
        )

        return {
            "input_ids": inputs["input_ids"].squeeze(0),
            "attention_mask": inputs["attention_mask"].squeeze(0),
            "labels": labels["input_ids"].squeeze(0)
        }


# =========================================================
# TRAINING CALLBACK
# =========================================================
class CustomCallback(TrainerCallback):
    def on_step_end(self, args, state, control, **kwargs):
        if state.global_step % args.logging_steps == 0 and state.log_history:
            loss = state.log_history[-1].get("loss", "N/A")
            print(f"Step {state.global_step} | Loss: {loss}")


# =========================================================
# METRIC PLOTS
# =========================================================
def plot_confusion_matrix(y_true, y_scores, model_name, threshold=0.5):
    y_pred = (np.array(y_scores) >= threshold).astype(int)

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=["Incorrect", "Correct"]
    )

    disp.plot(cmap="Blues", values_format="d")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.grid(False)
    plt.show()


def plot_roc_auc(y_true, y_scores, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr, linewidth=2, label=f"AUC = {roc_auc:.3f}")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_name}")
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

    return roc_auc


def generate_all_metrics(y_true, y_scores, model_name):
    print(f"\n📊 Metrics for {model_name}")
    plot_confusion_matrix(y_true, y_scores, model_name)
    auc_score = plot_roc_auc(y_true, y_scores, model_name)
    print(f"✅ AUC ({model_name}): {auc_score:.4f}")


# =========================================================
# QA SYSTEM
# =========================================================
class QASystem:
    def __init__(self, model_type):
        self.model_type = model_type

    # -----------------------------------------------------
    # ROBUST DATA LOADER (JSON + JSONL)
    # -----------------------------------------------------
    def load_and_sample_data(self, file_path):
        print(f"📂 Loading dataset: {file_path}")

        data = []
        with open(file_path, "r", encoding="utf-8") as f:
            first_char = f.read(1)
            f.seek(0)

            if first_char == "[":
                data = json.load(f)
            else:
                for line in f:
                    if line.strip():
                        data.append(json.loads(line))

        print(f"✓ Raw samples: {len(data)}")

        valid = []
        for ex in data:
            try:
                if (
                    "context" in ex and
                    "question" in ex and
                    "answers" in ex and
                    isinstance(ex["answers"], dict) and
                    "text" in ex["answers"] and
                    len(ex["answers"]["text"]) > 0
                ):
                    valid.append(ex)
            except Exception:
                continue

        print(f"✓ Valid QA samples: {len(valid)}")

        sampled = random.sample(valid, int(len(valid) * SAMPLE_FRACTION))

        train_data, val_data = train_test_split(
            sampled, test_size=0.2, random_state=42
        )

        print(f"✓ Train: {len(train_data)} | Validation: {len(val_data)}")
        return train_data, val_data

    # -----------------------------------------------------
    # MODEL INITIALIZATION
    # -----------------------------------------------------
    def initialize_model(self):
        tokenizer = T5TokenizerFast.from_pretrained("t5-small")
        model = T5ForConditionalGeneration.from_pretrained("t5-small")

        if self.model_type == "t5-lora":
            lora_cfg = LoraConfig(
                task_type=TaskType.SEQ_2_SEQ_LM,
                r=8,
                lora_alpha=32,
                lora_dropout=0.1,
                inference_mode=False
            )
            model = get_peft_model(model, lora_cfg)
            model.print_trainable_parameters()

        model.to(DEVICE)
        return model, tokenizer

    # -----------------------------------------------------
    # TRAIN
    # -----------------------------------------------------
    def train(self, model, tokenizer, train_data, val_data):
        train_ds = QADataset(tokenizer, train_data)
        val_ds = QADataset(tokenizer, val_data)

        args = Seq2SeqTrainingArguments(
            output_dir=f"./results_comparison/{self.model_type}",
            num_train_epochs=NUM_EPOCHS,
            per_device_train_batch_size=BATCH_SIZE,
            per_device_eval_batch_size=BATCH_SIZE,
            logging_steps=50,
            save_steps=500,
            report_to="none",
            predict_with_generate=True
        )

        trainer = Seq2SeqTrainer(
            model=model,
            args=args,
            train_dataset=train_ds,
            eval_dataset=val_ds,
            data_collator=default_data_collator,
            callbacks=[CustomCallback()]
        )

        tracker = EmissionsTracker(
            output_dir="emissions_data",
            output_file=f"{self.model_type}_train.csv"
        )

        tracker.start()
        trainer.train()
        emissions = tracker.stop()

        return emissions

    # -----------------------------------------------------
    # EVALUATE
    # -----------------------------------------------------
    def evaluate(self, model, tokenizer, val_data):
        model.eval()

        y_true, y_scores = [], []

        for qa in tqdm(val_data[:EVAL_SAMPLE_SIZE], desc="Evaluating"):
            input_text = f"question: {qa['question']} context: {qa['context']}"

            inputs = tokenizer(
                input_text,
                truncation=True,
                max_length=MAX_LENGTH,
                return_tensors="pt"
            ).to(DEVICE)

            with torch.no_grad():
                outputs = model.generate(**inputs, max_length=32)

            pred = tokenizer.decode(outputs[0], skip_special_tokens=True)
            gold = qa["answers"]["text"]

            em, f1 = self.compute_metrics(pred, gold)

            y_true.append(em)
            y_scores.append(f1)

        return y_true, y_scores

    # -----------------------------------------------------
    # METRICS
    # -----------------------------------------------------
    def compute_metrics(self, pred, golds):
        def normalize(text):
            text = text.lower()
            return "".join(c for c in text if c.isalnum() or c.isspace())

        pred = normalize(pred)
        golds = [normalize(g) for g in golds]

        em = int(pred in golds)

        pred_tokens = set(pred.split())
        best_f1 = 0.0

        for g in golds:
            g_tokens = set(g.split())
            common = pred_tokens & g_tokens
            if common:
                p = len(common) / len(pred_tokens)
                r = len(common) / len(g_tokens)
                f1 = 2 * p * r / (p + r + 1e-9)
                best_f1 = max(best_f1, f1)

        return em, best_f1


# =========================================================
# MAIN
# =========================================================
if __name__ == "__main__":

    DATA_PATH = "/content/adv_qa.json"

    for model_type in ["t5", "t5-lora"]:

        print(f"\n🚀 Running {model_type.upper()}")

        qa = QASystem(model_type)

        train_data, val_data = qa.load_and_sample_data(DATA_PATH)

        model, tokenizer = qa.initialize_model()

        train_emissions = qa.train(model, tokenizer, train_data, val_data)

        y_true, y_scores = qa.evaluate(model, tokenizer, val_data)

        generate_all_metrics(y_true, y_scores, model_type)

        print(f"🌱 Training CO₂ Emissions ({model_type}): {train_emissions:.4f} kg")
```
